### Project Setup

```python
""" Imports """
import kagglehub
from kagglehub import KaggleDatasetAdapter
from pathlib import Path
```

```python
""" Constants """
DATASET = "blastchar/telco-customer-churn"
DATASET_FILE = "WA_Fn-UseC_-Telco-Customer-Churn.csv"
BRONZE_DIR = Path("../bronze")
BRONZE_OUTPUT = BRONZE_DIR / "telco_churn_bronze.parquet"
```

## Bronze Layer

Dados brutos do dataset Telco Customer Churn do Kaggle.

```python
try:
    df = kagglehub.dataset_load(
        KaggleDatasetAdapter.PANDAS,
        DATASET,
        DATASET_FILE,
    )
except Exception as exc:
    raise RuntimeError(
        "Falha ao baixar/carregar dataset do Kaggle. Verifique suas credenciais "
        "(~/.kaggle/kaggle.json ou variáveis KAGGLE_USERNAME e KAGGLE_KEY)."
    ) from exc

print("Dataset carregado diretamente do Kaggle.")
print(f"Linhas: {len(df):,} | Colunas: {len(df.columns)}")
```

## Bronze Layer — Persistência

Salva o dado bruto em Parquet na camada Bronze. Nenhuma transformação é feita — o arquivo é uma cópia fiel do CSV original.

```python
BRONZE_DIR.mkdir(parents=True, exist_ok=True)
df.to_parquet(BRONZE_OUTPUT, index=False)

print(f"Bronze salvo em: {BRONZE_OUTPUT}")
print(f"Linhas: {len(df):,} | Colunas: {len(df.columns)}")
```
